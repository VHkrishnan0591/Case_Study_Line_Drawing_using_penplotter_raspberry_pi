```python
# Pasting the logo

from rembg import remove
input = cv2.imread("D:\EMBEDDED PRESENTATION\src\cat_image.jpg")
output = remove(input)
height, width = output.shape[:2]

# Create a white background (255, 255, 255 for RGB white)
background = np.ones((height, width, 3), dtype=np.uint8) * 255
rgb = output[:, :, :3]
alpha = (output[:, :, 3] / 255.0)  # Normalize alpha to 0-1 range

# Overlay the RGB array onto the white background using alpha transparency
for c in range(3):
    background[:, :, c] = (alpha * rgb[:, :, c] + (1 - alpha) * background[:, :, c]).astype(np.uint8)
cv2.imwrite('output_image_with_white_background.jpg', background)


image = cv2.imread("D:\EMBEDDED PRESENTATION\src\output_image_with_white_background.jpg")
logo = cv2.imread("D:\EMBEDDED PRESENTATION\src\THD_logo.jpg")
print(image.shape)
print(logo.shape)

image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
logo = cv2.resize(logo, None, fx=0.75, fy=0.75, interpolation=cv2.INTER_AREA)
h1, w1, _ = image.shape
h2, w2, _ = logo.shape
# Calculate the position to place the smaller image on the larger image
x_offset = w1 - w2
y_offset = 0

# Check if the smaller image fits within the larger image
# if x_offset < 0 or y_offset < 0:
#     print("Error: Smaller image doesn't fit on the top right corner.")
#     return

# Place the smaller image on the larger image
image[y_offset:y_offset+h2, x_offset:x_offset+w2] = logo
fig, ax = plt.subplots(figsize=(8, 8))
cv2.imwrite("PastedImage_with_white_bg.jpg",image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ax.imshow(image)
```

```python
image = cv2.imread("D:\EMBEDDED PRESENTATION\src\output_image_with_white_background.jpg")
logo = cv2.imread("D:\EMBEDDED PRESENTATION\src\THD_logo.jpg")
print(image.shape)
print(logo.shape)
image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
logo = cv2.resize(logo, None, fx=0.75, fy=0.75, interpolation=cv2.INTER_AREA)
h1, w1, _ = image.shape
h2, w2, _ = logo.shape
# Calculate the position to place the smaller image on the larger image
x_offset = w1 - w2
y_offset = 0
overlay = np.zeros((h1,w1,3),dtype='uint8')
overlay[y_offset:y_offset+h2, x_offset:x_offset+w2] = logo
image1 = cv2.addWeighted(overlay,1,image,0.3,0,image)
image2 = cv2.addWeighted(image,0.9,overlay,0.1,0,image)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(image1)
axs[1].imshow(image2)
axs[0].set_title('Image 1')
axs[1].set_title('Image 2')
plt.imshow(image)
```
